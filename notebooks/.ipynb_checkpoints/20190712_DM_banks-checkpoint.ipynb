```python
public_banks_df = pd.read_csv('../references/banks_tickers.txt', sep='\t',  skiprows = 1, names = ['Index', 'Name', 'Ticker'])
public_banks_df.head()
```

```python
#Fetch all stocks objects via APIs using token once
stock_list = []
for i in range(len(public_banks_df)):
    ticker = public_banks_df.Ticker.iloc[i]
    stock = Stock(ticker, output_format = 'pandas', token = key)
    stock_list.append(stock)
len(stock_list)
```
